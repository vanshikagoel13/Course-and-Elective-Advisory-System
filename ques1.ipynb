```python
import numpy as np
import numdifftools as nd
import matplotlib.pyplot as plt
```

```python
##  Problem 1 a)

# defining the given f(x) function
def f(x):
    return x**2 + np.log(x)

a = 1
#Linear Taylor Polynomial is of form: - L(x) = f(a) + (x-a)f'(a) and
#Quadratic Taylor Polynomial is of form: - Q(x) = f(a) + (x-a)*f'(a) + (((x-a)^2)*f''(a))/2 
#where a = 1 (given) and f'(x) is the first and f''(x) is the 2nd derivatives of f(x).

# for 1st derivative
der1 = nd.Derivative(f, full_output = True)
val1, info1 = der1(a)

#for 2nd derivative
der2 = nd.Derivative(f, n=2, full_output = True)
val2, info2 = der2(a)

# defining Linear Taylor Polynomial 
def L(x):
    return f(a) + val1*(x-a)

# defining Quadratic Taylor Polynomial 
def Q(x):
    return f(a) + val1*(x-a) + (val2*((x-a)**2))/2

x = np.linspace(0, 2, 500)
y1 = []
y2 = []
y3 = []
for i in x:
    y1.append(f(i))
    y2.append(L(i))
    y3.append(Q(i))

plt.plot(x,y1,label = "f(x)")
plt.plot(x,y2,label = "L(x)")
plt.plot(x,y3,label = "Q(x)")
plt.title('Linear and Quadratic Taylor approximations for f(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()


```

```python
# Problem 1 b)

# eL(x) = f(x)-L(x)
# eQ(x) = f(x)-Q(x)
yel = []
yeq = []

def errL(x):
    return (f(x)-L(x))/(x-1)

def errQ(x):
    return (f(x)-Q(x))/((x-1)**2)

for i in x:
    yel.append(errL(i))
    yeq.append(errQ(i))

plt.plot(x,yel,label = "error in linear")
plt.plot(x,yeq,label = "error in quadratic")
plt.title('error associated with approximations')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

```
